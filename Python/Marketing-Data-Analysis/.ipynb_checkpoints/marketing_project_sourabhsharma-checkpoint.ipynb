## Outlier detection and Treatment (IQR)

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

numerical_columns = ['Income', 'TotalSpending', 'TotalPurchases']
outliers_dict = {}
for col in numerical_columns:
    outliers_dict[col], marketing_df = detect_outliers(marketing_df, col)

# Print outliers for verification
for col, outliers in outliers_dict.items():
    print(f"Outliers in {col}:")
    print(outliers[[col]])
```

## Hypothesis testing

```python
younger_group=marketing_df[marketing_df['Age']<50]['NumStorePurchases'].value_counts().reset_index()
older_group=marketing_df[marketing_df['Age']>50]['NumStorePurchases'].value_counts().reset_index()
```

```python
younger_group['count'].sum()
```

```python
older_group['count'].sum()
```

```python
#Hypothesis Testing:
# Older individuals and in-store shopping preference

younger_group=marketing_df[marketing_df['Age']<50]['NumStorePurchases']
older_group=marketing_df[marketing_df['Age']>50]['NumStorePurchases']
t_stat, p_value = ttest_ind(younger_group, older_group, nan_policy='omit', equal_var=False)

print("T-test for Store Purchases by Age:")
print(f"T-statistic: {t_stat}, P-value: {p_value}")
    
# Interpreting the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in store purchases between younger and older individuals.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in store purchases between younger and older individuals.")
```

```python
## Data Projection
younger_counts = marketing_df[marketing_df['Age'] < 50]['NumStorePurchases'].value_counts().sort_index()
older_counts = marketing_df[marketing_df['Age'] >= 50]['NumStorePurchases'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
plt.bar(younger_counts.index, younger_counts.values, alpha=0.8, label='Under 50')
plt.bar(older_counts.index, older_counts.values, alpha=0.6, label='50 and Above')
plt.xlabel("Number of Store Purchases")
plt.ylabel("Count")
plt.title("Distribution of Store Purchases by Age Group")
plt.legend()
plt.show()
```

```python
# (B) Customers with children and online shopping preference
with_kids=marketing_df[marketing_df['TotalKids']>=1]['NumStorePurchases']
without_kids=marketing_df[marketing_df['TotalKids']==0]['NumStorePurchases']

t_test, p_value=ttest_ind(with_kids, without_kids)
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in online shopping orders for people with kids and without kids.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in online purchases by people with or without kids.")
```

```python
## Data Projection
with_kids = marketing_df[marketing_df['TotalKids']>=1]['NumStorePurchases'].value_counts().sort_index()
older_counts = without_kids=marketing_df[marketing_df['TotalKids']==0]['NumStorePurchases'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
plt.bar(with_kids.index, with_kids.values, alpha=0.6, label='With Kids')
plt.bar(without_kids.index, without_kids.values, alpha=1, label='Without Kids')
plt.xlabel("Number of online purchases")
plt.ylabel("Count")
plt.title("Online shopping preference")
plt.legend()
plt.show()
```

## Does the United States significantly outperform the rest of the world in total purchase volumes?


```python
marketing_df.groupby([marketing_df['Country_US']])['TotalPurchases'].sum()
```

```python
US_purchases=marketing_df[marketing_df['Country_US']==1]['TotalPurchases']
Other_purchases=marketing_df[marketing_df['Country_US']==0]['TotalPurchases']
print("ANOVA test for total Purchases US vs Rest" , f_oneway(US_purchases, Other_purchases))
anova_result=f_oneway(US_purchases, Other_purchases)
# Explanation of the results
if anova_result.pvalue < 0.05:
    print("The p-value is less than 0.05, which suggests that there is a significant difference in total purchases between customers from the US and those from other countries.")
else:
    print("The p-value is greater than 0.05, indicating that there is not enough evidence to say that the average total purchases are different between customers from the US and those from other countries.")
    
print("In simple terms, this means that the purchasing behavior appears to be similar across these two groups based on the data analyzed.")
```
